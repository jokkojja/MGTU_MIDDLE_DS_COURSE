```python
# test processing

def preprocess(text):
    text = text.lower()
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', text) #change url on URL
    text = re.sub('@[^\s]+', 'USER', text) # change @user on USER # letters and numbers
    text = re.sub(' +', ' ', text)
    return text.strip()
```
