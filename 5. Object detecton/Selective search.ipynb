```python
import numpy as np
import matplotlib.pyplot as plt
import os
import datetime
import cv2
import sys
import tensorflow as tf
```

## Upload data

```python
im = cv2.imread('full-158ad650896fbbeb4931f32ad54de833.jpeg')
```

```python
new_height = 300
new_width = int(im.shape[1] * new_height / im.shape[0]) #proportional width
```

```python
im = cv2.resize(im, (new_height, new_width))
```

## Selective search

```python
# multy processs
cv2.setUseOptimized(True)
cv2.setNumThreads(8)
```

```python
sss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
```

```python
sss.setBaseImage(im)
```

```python
sss.switchToSelectiveSearchQuality()
```

```python
rects = sss.process()
```

```python
plt.figure(figsize = (10,10))
plt.imshow(im)
```

## Image processing

```python
num_show_rects = 10 # count of region proposal

im_out = im.copy()

for i, rect in enumerate(rects):
    if i < num_show_rects:
        x,y,w,h = rect
        cv2.rectangle(im_out, (x,y), (x+w, y+w), (0,255,0), 1, cv2.LINE_AA)
    else:
        break

plt.figure(figsize = (10,10))
plt.imshow(im_out)
```
